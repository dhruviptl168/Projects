**Simple Game Score Tracker**


```python
score = 0

def ask_question(question, answer):
    global score
    user_answer = input(question + " ")
    if user_answer.lower() == answer.lower():
        print("Correct Answer!")
        score += 10
    else:
        print("Wrong Answer!")

print("Current Score:", score)
print("Simple Quiz Game")

ask_question("1. What is the capital of India?", "delhi")
ask_question("2. What is 5 + 4?", "9")
ask_question("3. What color is the sky?", "blue")

print("Game Over!")
print("Final Score:", score)
```
