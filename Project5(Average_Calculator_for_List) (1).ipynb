**Average Calculator for List**

```python
def get_numbers():
    numbers = []
    n = int(input("Enter how many numbers you want: "))
    for i in range(n):
        num = float(input("Enter number: "))
        numbers.append(num)
    return numbers
def sum(num_list):
    total = 0
    for num in num_list:
        total = total + num
    return total
def average(num_list):
    total = sum(num_list)
    avg = total / len(num_list)
    return avg
def display(numbers, total, avg):
    print(numbers)
    print(total)
    print(avg)



numbers = get_numbers()
total = sum(numbers)
average = average(numbers)
display(numbers, total, average)
```
