
**List Sum Calculator**

```python
def get_numbers():
    numbers = []
    n = int(input("Enter how many numbers you want to add: "))

    for i in range(n):
        num = int(input("Enter number {}: ".format(i+1)))
        numbers.append(num)
    return numbers
def calculate_sum(numbers):
    total = 0
    for i in numbers:
        total = total + i
    return total
def display_result(numbers, total):
    print("\nEntered List:", numbers)
    print("Total Sum =", total)
def main():
    print("List Sum Calculator")

    numbers = get_numbers()
    total = calculate_sum(numbers)
    display_result(numbers, total)


main()
```
