**Car Information System**

```python
class Car:
    def __init__(self, brand, model, speed):
        self.brand = brand
        self.model = model
        self.speed = speed
    def display_info(self):
        print("Car Details")
        print(self.brand)
        print(self.model)
        print(self.speed, "km/h")


cars = []
n = int(input("Enter number of cars: "))
for i in range(n):
    print("Enter details of Car", i+1)

    brand = input("Enter Car Brand: ")
    model = input("Enter Car Model: ")
    speed = int(input("Enter Car Speed: "))


    car = Car(brand, model, speed)
    cars.append(car)

print("Car Information")

for car in cars:
    car.display_info()
```
